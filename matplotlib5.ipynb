Matplotlib-grafiikassa lukuisilla parametreilla on oletusarvot, joita voi muuttaa. Jos esimerkiksi haluat käyttää kaavioissa tiettyä fonttia, niin ei kannata vaihtaa fonttia jokaiselle kaaviolle erikseen. Oletusparametrit ja niiden arvot voit listata `rcParams`-käskyllä.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-white')

from matplotlib import rcParams

rcParams
```

## Fontit ja fonttikoot

Seuraavassa määrään kaavioissa käytettäväksi fontiksi pääteviivattoman fontin (**sans-serif**) ja tarkemmin fonttilajiksi **Arial**. Lisäksi määritän kaavion eri osien fonttien kokoja. Lopuksi määritään kuvion oletuskooksi 12 x 8. Määritykset ovat voimassa kaikissa tämän jälkeen tähän muistioon tehtävissä kaavioissa.

```python
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
rcParams['font.size'] = 12 
rcParams['axes.titlesize'] = 24
rcParams['axes.labelsize'] = 20
rcParams['xtick.labelsize'] = 14
rcParams['ytick.labelsize'] = 14
rcParams['legend.fontsize'] = 14
rcParams['figure.figsize'] = [12, 8]
```

```python
df = pd.read_excel('http://taanila.fi/data1.xlsx')

df1 = pd.crosstab(df['koulutus'], 'f')

df1.plot.barh(legend=False)

plt.title('Koulutusjakauma')
plt.xlabel('Lukumäärä')

plt.grid(axis='x')

for bar in plt.gca().patches:
    x = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    plt.text(x, y, f' {x:.0f} %', va='center')
```
