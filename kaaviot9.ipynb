<h2>Aikasarjojen vertailu</h2>

```python
#luon dataframen
df = pd.DataFrame({'Etelä':[120,130,167,210,130,135,172,213,134],
                'Keski':[98,102,120,180,103,109,140,189,104],
                'Pohjoinen':[89,95,108,120,88,99,110,123,88]},
                index=['1.2009','2.2009','3.2009','4.2009','1.2010','2.2010','3.2010','4.2010','1.2011'])
df
```

Huomaa, miten indeksin arvot tulevat vaaka-akselille.

Lisäparametri <b>style</b> määrittää viivan ja havaintopisteiden tyylin. Lisätietoa: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

```python
ax1 = df.plot(style='-o', rot=45)
```
